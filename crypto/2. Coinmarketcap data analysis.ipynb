```python
import pandas as pd
import numpy as np
```

```python
cmc = pd.read_csv('data/cmc/cmc_20210217.csv')
```

```python
cmc.head()
```

```python
# Let's check the ratio between market cap and volume_24h
# we will assign the symbol mc/v which implies market cap / volume

cmc['mc_by_v']=cmc.market_cap/cmc.volume_24h
```

# Filter 1:
    We need a high volatile crypto !!
    So we select those whose 24 hour volume is equal or more that market cap! Thus mc_by_v value should be less than or equal to 1 

```python
cmc=cmc[cmc['mc_by_v']<=1]
```

# Filter 2: 
    We need to select those coin with market_cap of atlease 100 million USD

```python
cmc=cmc[cmc['market_cap']>=100000000]
```
